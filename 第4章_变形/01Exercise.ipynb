```python
table_m = table[['ID','Gender','Math']]
table_m.head()
```

```python
table.pivot(index='ID',columns='Gender',values='Math').head()
```

```python
pivoted = table.pivot(index='ID',columns='Gender',values='Math')
result = pivoted.reset_index().melt(id_vars=['ID'],value_vars=['F','M'],value_name='Math')\
                     .dropna().set_index('ID').sort_index()
result.equals(table_m.set_index('ID'))
```

```python
table_s = pd.pivot_table(table,index=['Class','ID'],columns='Gender',values=['Height','Weight'])
table_s.groupby('Class').head(2)
```

```python
table_stacked = table_s.stack()
table_stacked.groupby('Class').head(2)
```

```python
table_stacked = table_s.stack(0)
table_stacked.groupby('Class').head(2)
```

```python
table_stacked.head()
```

```python
result = table_stacked.unstack().swaplevel(1,0,axis=1).sort_index(axis=1)
result.equals(table_s)
```
